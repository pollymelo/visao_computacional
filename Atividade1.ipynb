# Parte 3: Transformações de Imagens

**Redução da resolução da imagem**

Com a resolução da imagem à 50%, continuamos a identificar as frutas, mas perdemos detalhes como textura e variação de cor

Com a resolução da imagem à 20%, continuamos a identificar as frutas, mas ela está pixelada.

```python
maca = {
    'Maca1.jpeg': 'Maçã Pão de Açúcar',
    'Maca2.jpeg': 'Maçã Super Benfica (Benfica)',
    'Maca3.jpeg': 'Maçã Super Benfica (Benfica)',
    'Maca4.jpeg': 'Maçã Cometa (Benfica)',
    'Maca5.jpeg': 'Maçã Super Benfica (José Bonifácio)'
}

for filename, title_prefix in maca.items():
    img = cv2.imread(filename)

    if img is None:
        print(f"Error: Could not load image {filename}")
        continue

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    altura, largura = img.shape[:2]

    img_50 = cv2.resize(img, (largura // 2, altura // 2))
    img_10 = cv2.resize(img, (largura // 10, altura // 10))

    fig, ax = plt.subplots(1, 3, figsize=(15, 5))

    ax[0].imshow(img)
    ax[0].set_title(f"Original - {title_prefix}")

    ax[1].imshow(img_50)
    ax[1].set_title("50% Size")

    ax[2].imshow(img_10)
    ax[2].set_title("10% Size")

    for a in ax:
        a.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
laranja = {
    'Laranja1.jpeg': 'Maçã Super Benfica (Benfica)',
    'Laranja2.jpeg': 'Maçã Super Benfica (Benfica)',
    'Laranja3.jpeg': 'Maçã Cometa (Benfica)',
    'Laranja4.jpeg': 'Maçã Cometa (Benfica)',
    'Laranja5.jpeg': 'Maçã Super Benfica (José Bonifácio)'
}

for filename, title_prefix in laranja.items():
    img = cv2.imread(filename)

    if img is None:
        print(f"Error: Could not load image {filename}")
        continue

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    altura, largura = img.shape[:2]

    img_50 = cv2.resize(img, (largura // 2, altura // 2))
    img_10 = cv2.resize(img, (largura // 10, altura // 10))

    fig, ax = plt.subplots(1, 3, figsize=(15, 5))

    ax[0].imshow(img)
    ax[0].set_title(f"Original - {title_prefix}")

    ax[1].imshow(img_50)
    ax[1].set_title("50% Size")

    ax[2].imshow(img_10)
    ax[2].set_title("10% Size")

    for a in ax:
        a.axis('off')

    plt.tight_layout()
    plt.show()
```

**Parte B - Espaço de Cor**

Conversões:
*   RGB
*   HSV
*   Escala de Cinza

**Diferenças visuais observadas**

RGB:
*   Mantém as cores das imagens
*   As cores continuam vibrantes

HSV:
*   Identificação das áreas vibrantes das cores

Escala de cinza:
*   Perca da informação cromática

**Aplicações**

RGB:
*   Classificação de objetos


HSV:
*   Detecção de cores

Escala de cinza:
*   Redução do custo computacional

```python
for filename, title_prefix in maca.items():
    img_bgr = cv2.imread(filename)

    if img_bgr is None:
        print(f"Error: Could not load image {filename}")
        continue

    rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)
    gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

    fig, ax = plt.subplots(1, 3, figsize=(18, 5))

    ax[0].imshow(rgb)
    ax[0].set_title(f"RGB - {title_prefix}")

    ax[1].imshow(hsv)
    ax[1].set_title(f"HSV - {title_prefix}")

    ax[2].imshow(gray, cmap='gray')
    ax[2].set_title(f"Escala de Cinza - {title_prefix}")

    for a in ax:
        a.axis('off')

    plt.tight_layout()
    plt.show()
```

**Parte C - Quantização**

Níveis utilizados:

*   256 níveis
*   64 níveis
*   32 níveis
*   2 níveis

**Observações**

A quantização controla a precisão tonal da imagem. Com isso, quanto menos os bits da imagem, percebemos que há uma maior chance de erro e perca de contraste.



```python
for filename, title_prefix in maca.items():
    img_gray = cv2.imread(filename, 0)

    if img_gray is None:
        print(f"Error: Could not load image {filename}")
        continue

    # 64 níveis
    q64 = (img_gray // 4) * 4
    # 32 níveis
    q32 = (img_gray // 8) * 8
    # 2 níveis
    q2 = ((img_gray > 127) * 255).astype(np.uint8)

    fig, ax = plt.subplots(1, 4, figsize=(20, 5))

    ax[0].imshow(img_gray, cmap='gray')
    ax[0].set_title(f"Original (256 níveis) - {title_prefix}")

    ax[1].imshow(q64, cmap='gray')
    ax[1].set_title("64 níveis")

    ax[2].imshow(q32, cmap='gray')
    ax[2].set_title("32 níveis")

    ax[3].imshow(q2, cmap='gray')
    ax[3].set_title("2 níveis")

    for a in ax:
        a.axis('off')

    plt.tight_layout()
    plt.show()
```

**Parte D - Formato de Arquivo**

Formatos:

*   JPEG com compressão
*   PNG sem perdas

**Observação**

A câmera do celular possui uma boa qualidade que não se faz tão sentida ao reduzir o tamanho do arquivo das imagens. Apenas há uma perca de brilho e saturação que não faz que a imagem perca a sua classificação.

```python
for filename, title_prefix in maca.items():
    img = cv2.imread(filename)

    if img is None:
        print(f"Error: Could not load image {filename}")
        continue

    base_name = os.path.splitext(filename)[0]

    jpeg_output_filename = f'{base_name}_compressao.jpg'
    cv2.imwrite(jpeg_output_filename, img, [cv2.IMWRITE_JPEG_QUALITY, 30])

    png_output_filename = f'{base_name}_sem_perda.png'
    cv2.imwrite(png_output_filename, img)

    jpg_size = os.path.getsize(jpeg_output_filename) / 1024
    png_size = os.path.getsize(png_output_filename) / 1024

    print(f"--- {title_prefix} ({filename}) ---")
    print(f"Tamanho JPEG (qualidade 30): {jpg_size:.2f} KB")
    print(f"Tamanho PNG (sem perdas): {png_size:.2f} KB")
    print("\n")
```

# Parte 4: Estrutura do mini dataset

```python
import os
import shutil

dataset_base_dir = 'dataset'

class_a_dir = os.path.join(dataset_base_dir, 'classe_A')
class_b_dir = os.path.join(dataset_base_dir, 'classe_B')

os.makedirs(class_a_dir, exist_ok=True)
os.makedirs(class_b_dir, exist_ok=True)

print(f"Created directory: {class_a_dir}")
print(f"Created directory: {class_b_dir}")
```

```python
img_counter_a = 1
for original_filename, _ in maca.items():
    new_filename = f'img{img_counter_a:03d}_original.jpg'
    destination_path = os.path.join(class_a_dir, new_filename)

    shutil.copy(original_filename, destination_path)

    print(f"Copied '{original_filename}' to '{destination_path}'")
    img_counter_a += 1

img_counter_b = 6
for original_filename, _ in laranja.items():
    new_filename = f'img{img_counter_b:03d}_original.jpg'
    destination_path = os.path.join(class_b_dir, new_filename)

    shutil.copy(original_filename, destination_path)

    print(f"Copied '{original_filename}' to '{destination_path}'")
    img_counter_b += 1

print("Dataset organization complete!")
```
